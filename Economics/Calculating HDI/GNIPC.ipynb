```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize':(20,10)})
```

```python
df = pd.read_csv('data/GNIPC.csv', encoding='latin1')
```

```python
df.head()
```

```python
df['gni_ppp_difference'] = df.gni - df.ppp
sns.lmplot(
    x='gni',
    y='ppp',
    data=df.loc[(df.gni.notna()) & (df.ppp.notna())]
)

plt.title('GNI vs. PPP')
```

```python
columns = ['economy', 'gni_ranking', 'gni', 'ppp_ranking', 'ppp']
df.sort_values('gni_ranking').loc[0:9, columns]
```
