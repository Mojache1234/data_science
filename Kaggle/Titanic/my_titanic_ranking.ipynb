```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/titanic-publicleaderboard.csv')
df.head()
```

```python
# df['Score_normalized'] = (df.Score - df.Score.mean()) / df.Score.std()
# sns.distplot(df.Score_normalized)
# sns.distplot(df.loc[df.Score >= 0.6].Score)
lower, upper = df.Score.quantile([0.25, 0.75])
IQR = (upper - lower)
lower_outliers = lower - IQR
upper_outliers = upper + IQR

scores = df.loc[(df.Score >= lower_outliers) & (df.Score <= upper_outliers)].Score
sns.distplot(scores)
plt.axvline(0.77990, color='r', label='Test')
plt.title('Ariel\'s Percentile Rank')
```

```python
sns.distplot((scores - scores.mean()) / scores.std())
plt.axvline(0.77990, 0, 1.7)
```
